```python
arr = [49, 12, 100, 276, 33]
n = 27
```

```python
def solution(arr, n):
    if len(arr)%2:
        for i in range(0, len(arr), 2):
            arr[i] += n
        return arr
    else:
        for i in range(1, len(arr), 2):
            arr[i] += n
        return arr
```

```python
def solution(arr, n):
    return  [arr[i] if i%2==1 else arr[i]+n for i in range(len(arr))] if len(arr)%2 else [arr[i]+n if i%2 else arr[i] for i in range(len(arr))]
```

```python
solution(arr, n)
```
