```python
def solution(array, n):
    array = sorted(array + [n])

    if array[array.index(n)] == array[-1]:
        return array[-2]
    elif array[array.index(n)] == array[0]:
        return array[1]
    else:
        pre_cost = array[array.index(n)] - array[array.index(n)-1] 
        post_cost = array[array.index(n)+1] - array[array.index(n)]
        if pre_cost <= post_cost:
            return array[array.index(n)-1] 
        else:
            return array[array.index(n)+1]
```

```python
array = [5,10,22,26]
n = 24
```

```python
solution(array, n)
```
