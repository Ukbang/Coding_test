```python
def solution(s):
    split_s = s.split(' ')
    while 'Z' in split_s:
        split_s.pop(split_s.index('Z')-1), split_s.pop(split_s.index('Z'))
    return eval('+'.join(split_s)) if len(split_s) >= 1 else 0
```

```python
def solution(s):
    split_s = s.split(' ')
    while 'Z' in split_s:
        del split_s[split_s.index('Z')-1]
        del split_s[split_s.index('Z')]
    return eval('+'.join(split_s)) if len(split_s) >= 1 else 0
```
