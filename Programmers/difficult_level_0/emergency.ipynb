```python
emergency = [5,2,18,4,7]
```

```python
def solution(emergency):
    _copy = sorted(emergency.copy(), reverse=True)
    for i,v in enumerate(_copy, start=1):
        emergency[emergency.index(v)] = [emergency[emergency.index(v)],i]

    return [i[1] for i in emergency]
```

```python
solution(emergency)
```
