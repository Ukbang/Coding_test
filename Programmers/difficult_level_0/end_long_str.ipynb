```python
def solution(my_string, pat):
    my_string_split = my_string.split(pat)
    if my_string_split.count('') > 1:
        result = list(map(lambda x:len(x), my_string))
        return my_string_split[result.index(max(result))]+pat
    else:
        return my_string_split[my_string_split.index('')-1]+pat
```
