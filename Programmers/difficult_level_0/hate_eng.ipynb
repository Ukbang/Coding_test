```python
def solution(numbers):
    num_dict = {'zero':'0', 'one':'1', 'two':'2', 'three':'3', 'four':'4', 'five':'5', 'six':'6', 'seven':'7', 'eight':'8', 'nine':'9'}

    for dic in num_dict:
        numbers = numbers.replace(dic, num_dict[dic])

    return int(numbers)
```

```python
numbers = "onetwothreefourfivesixseveneightnine"
```

```python
solution(numbers)
```
