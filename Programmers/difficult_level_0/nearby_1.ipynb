```python
arr = [1, 1, 1, 1, 0]
idx = 3
```

```python
def solution(arr, idx): # 오히려 조건식보다 빠름
    for i,v in enumerate(arr[idx:]):
        if v==1:
            return i+idx
    return -1 
```

```python
def solution(arr, idx): # 시간복잡도 최소화
    result = [i+idx for i,v in enumerate(arr[idx:]) if v==1]
    return -1 if result==[] else result[0]
```
