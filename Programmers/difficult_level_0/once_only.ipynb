```python
s = 'abcabcadc'
```

```python
def solution(s):
    once_dict = {}
    for _s in list(s):
        if _s in once_dict: 
            once_dict[_s] += 1
        else:
            once_dict[_s] = 1

    return ''.join(sorted([i for i,v in once_dict.items() if v==1]))
```

```python
solution(s)
```
