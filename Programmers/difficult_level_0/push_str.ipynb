```python
def solution(A,B):
    A = list(A)
    result = 0
    loop = False
    for i in range(len(A)):
        if A == list(B):
            result = i
            loop = True
            break
        else:            
            A.insert(0, A.pop())

    return result if loop==True else -1
```

```python
solution('apple', 'elppa')
```
