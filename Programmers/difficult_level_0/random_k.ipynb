```python
arr = [0, 1, 1, 2, 2, 3]
k = 3
```

```python
def solution(arr, k):
    result = []
    for i in arr:
        if i not in result:
            result.append(i)
    if len(result) < k:
        return result + [-1]*(k-len(result))
    elif len(result) > k:
        return result[:k]
    else:
        return result
```

```python
solution(arr,k)
```
