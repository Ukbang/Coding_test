```python
def solution(k, score):
    k_score = []
    final_score = []
    for s in score:
        k_score.append(s)
        k_score = sorted(k_score)
        if len(k_score) > k:
            final_score.append(k_score[-k:][0])
        else:
            final_score.append(k_score[0])
    return final_score
```
