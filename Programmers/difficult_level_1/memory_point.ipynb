```python
name = ["may", "kein", "kain", "radi"]
yearning = [5, 10, 1, 3]
photo = [["may", "kein", "kain", "radi"],["may", "kein", "brin", "deny"], ["kon", "kain", "may", "coni"]]
```

```python
def solution(name, yearning, photo):
    name_yearning = {}
    result = []
    for i,v in zip(name,yearning):
        name_yearning[i]= v
    
    for _photo in photo:
        memory = 0
        for p in _photo:
            try:
                memory += name_yearning[p]
            except:
                memory += 0
        result.append(memory)

    return result
```

```python
solution(name, yearning, photo)
```
