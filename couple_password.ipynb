```python
def solution(s, skip, index):
    alphabet = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']

    input = [i for i in s]
    for word in skip:
        del alphabet[alphabet.index(word)]

    result = []
    for i in input:
        position = alphabet.index(i)+index
        while position >= len(alphabet): 
            position -= len(alphabet)
        result.append(alphabet[position])

    return ''.join(result)
```
