```python
def solution(myStr):
    for s in ['a','b','c']:
       myStr = myStr.replace(s, ' ')

    result = []
    for i in myStr.split(' '):
        if i != '':
            result.append(i)       
    
    return ['EMPTY'] if result == [] else result
```
