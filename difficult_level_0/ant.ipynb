```python
hp = 54
```

```python
def solution(hp):
    result = 0
    while True:
        if hp == 0:
            return result
        elif hp >= 5:
            hp -= 5
            result += 1
        elif hp >= 3:
            hp -= 3
            result += 1
        else:
            hp -= 1
            result += 1
```

```python
solution(hp)
```
