```python
n = 10
```

```python
def solution(n):
    result = [n]
    while True:
        if n <= 1:
            return result
        else:
            if n % 2:
                n = 3 * n + 1
                result.append(n)
            else:
                n //= 2
                result.append(n)
        
```

```python
solution(n)
```
