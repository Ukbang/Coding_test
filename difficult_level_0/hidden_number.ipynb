```python
my_string = "kingof98i1k33a2"
```

```python
def solution(my_string):
    result = []
    for i,s in enumerate(my_string):
        if i==0 and s.isdigit():
            result.append(s)
        elif s.isdigit():
            if my_string[i-1].isdigit():
                result[-1] += s
            else:
                result += [s]
    
    return 0 if len(result)==0 else sum(map(int, result))
```

```python
solution(my_string)
```
