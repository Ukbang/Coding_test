# 럭키 스트레이트!




lucky.jpg

```python
score = input()

mean = len(score)//2
score = list(score)
left_score = score[:mean]
right_score = score[mean:]

left_total = 0
right_total = 0

if len(left_score) != len(right_score):
    print('잘못된 입력입니다.')
    print('입력 예시: 123402 -> 123,402 처럼 좌 우의 길이가 같을 수 있도록 해주십시오.')
else:
    for num in range(mean):
        left_total += int(left_score[num])
        right_total += int(right_score[num])
        
    if left_total - right_total == 0:
        print('Lucky')
    else:
        print('Ready')    
```
