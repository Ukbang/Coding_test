```python
def solution(polynomial):
    x_number = 0
    number = 0
    if polynomial == 'x':
        return 'x'
    elif 'x' not in polynomial:
        try:
            _split = polynomial.split(' + ')
            return str(sum(map(int, _split)))
        except:
            return str(polynomial)
    else:
        _split = polynomial.split(' + ')
        for num in _split:
            if 'x' in num:
                num = num.replace('x', '')
                if num == '':
                    num = num.replace('', '1')
                x_number += int(num)
            else:
                number += int(num)
                
        if x_number == 1:
            return 'x + ' + str(number) if number != 0 else 'x'
    
        return str(x_number)+'x + ' + str(number) if number != 0 else str(x_number)+'x'
```
