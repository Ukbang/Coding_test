```python
def solution(num, total):
    x = total//num
    y = [x]
    x_m = x
    x_p = x
    x_t = x
    while True:
        if num < 1 or num > 100 or total < 0 or total > 1000: 
            print('제한된 범위가 아닙니다.')
            break
        else:
            x_p +=  1
            y.append(x_p)
            x_t +=  x_p
            if x_t == total:
                break
            else:
                x_m -=  1
                y.append(x_m)
                x_t += x_m
                if x_t == total:
                    break
                else:
                    continue
    answer = sorted(y)
    return answer
```
