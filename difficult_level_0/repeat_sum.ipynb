```python
def solution(num, total):
    mean = total // num
    min = mean
    max = mean
    result = [mean]
    loop = True
    while len(result) < num:
        if loop:
            max += 1
            result.append(max)
            loop = False
        else:
            min -= 1
            result.insert(0, min)
            loop = True

    return result
```

```python
num = 5
total = 5

solution(num, total)
```
