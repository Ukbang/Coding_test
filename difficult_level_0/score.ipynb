```python
score = [[80, 70], [70, 80], [30, 50], [90, 100], [100, 90], [100, 100], [10, 30]]
```

```python
def rank_match(test_score):
    for i in range(len(test_score)):
        test_score[i] = [test_score[i], i+1]

    for i in range(len(test_score)-1):
        if test_score[i][0] == test_score[i+1][0]:
            test_score[i+1][1] = test_score[i][1]

    return test_score

def solution(score):
    sub_score = [sum(i) for i in score]
    test_score = sorted(sub_score, reverse=True)

    test_score = rank_match(test_score)

    for i in range(len(test_score)):
        sub_score[sub_score.index(test_score[i][0])] = [sub_score[sub_score.index(test_score[i][0])]] + [test_score[i][1]]

    return [s[1] for s in sub_score]
    
```

```python
solution(score)
```
