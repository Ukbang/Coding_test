```python
left = 13
right = 17
```

```python
def solution(left, right):
    result = 0
    for i in range(left, right+1):
        divisor = 0
        for j in range(1, right+1):
            if j <= i and i%j==0:
                divisor += 1
        if divisor%2:
            result -= i
        else:
            result += i
    return result
```
