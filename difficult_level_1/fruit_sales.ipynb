```python
k = 4
m = 3
score = [4, 1, 2, 2, 4, 4, 4, 4, 1, 2, 4, 2]
```

```python
def solution(k, m, score): # 성공
    result = 0
    fruit = sorted(score, reverse=True)
    for i in range(m, len(score)+1, m):
        result += min(fruit[i-m:i])*m
    return result
```

```python
def solution(k, m, score): # 시간복잡도 실패
    return sum([min(sorted(score, reverse=True)[i-m:i])*m for i in range(m, len(score)+1, m)])
```
