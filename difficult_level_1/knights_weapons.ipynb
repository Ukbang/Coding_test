```python
def get_len_divisor(n):
    result = []
    for i in range(1, int(n**0.5)+1):
        if n % i == 0:
            result.append(i)
            if (i**2) != n:
                result.append(n // i)
    return len(result)

def solution(number, limit, power):
    return sum(list(map(lambda x : power if limit < x  else x, [get_len_divisor(knight) for knight in range(1, number+1)])))
```
