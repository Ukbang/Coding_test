```python
s = "try hello world"
```

```python
def solution(s):
    result = ""
    idx = 0
    for i in s:
        if i == " ":
            result += i
            idx = 0
        elif idx==1:
            result += i.lower()
            idx = 0
        elif idx==0:
            result += i.upper()
            idx = 1
    return result
```

```python
solution(s)
```
