```python
# 최종 완성본
result = [0]*20
score_dict = {'1':0, '2':1, '3':3, '4':5, '5':7, '6':9, '7':11, '8':15, '9':20, '10':25, '11':30,
              '12':35, '13':40, '14':50, '15':60, '16':70, '17':85, '18':100, '19': 150, '20':300}

def first_input(input):
    if input <= 10:
        below_10 = int(round(input/2, 0))

        if result[below_10] == 0:
            result[below_10] = input
            return print(result)
        
        elif input in result:
            for i in range(1, len(result)):
                if result[result.index(input)+i] == 0:
                    result[result.index(input)+i] = input
                    return print(result)   
                elif result[result.index(input)-i] == 0:
                    result[result.index(input)-i] = input
                    return print(result)
                result[result.index(0)] = input
                return print(result)
            
        elif result[below_10] != 0:
            if result[below_10] < input:
                for i in range(1, len(result)):
                    if result[below_10+i] == 0:
                        result[below_10+i] = input
                        return print(result)
            else:
                for i in range(1, len(result)):
                    if result[below_10-i] == 0:
                        result[below_10-i] = input
                        return print(result)                

    else:
        above_10 = int(round(input/1.5, 0))-1

        if result[above_10] == 0:
            result[above_10] = input
            return print(result)        
        
        elif input in result:
            for i in range(1, len(result)):
                if result[result.index(input)+i] == 0:
                    result[result.index(input)+i] = input
                    return print(result)   
                elif result[result.index(input)-i] == 0:
                    result[result.index(input)-i] = input
                    return print(result)
                result[result.index(0)] = input
                return print(result)            

        elif result[above_10] != 0:
            if result[above_10] < input:
                for i in range(1, len(result)):
                    if result[above_10+i] == 0:
                        result[above_10+i] = input
                        return print(result)
                    
            else:
                for i in range(1, len(result)):
                    if result[above_10-i] == 0:
                        result[above_10-i] = input
                        return print(result)                

# result에서 0보다 큰 값만 추출
def filtered(num):
    return num > 0

# 시작
def start():
    while len(list(filter(filtered, result))) < 20:
        first_input(int(input('숫자를 뽑아주세요.')))
    point()

def point():
    up = 1
    total_score = []
    for i in range(len(result)-1):
        if result[i] <= result[i+1]:
            up += 1
        else:
            total_score.append(score_dict[str(up)])
            up = 1
    total_score.append(score_dict[str(up)])
    return print(f"총 획득한 점수는 {total_score} 이며 합산 점수는 {sum(total_score)}점입니다.")

start()
```
